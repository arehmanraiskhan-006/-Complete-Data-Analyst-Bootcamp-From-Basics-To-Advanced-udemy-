#### Declaring and Assigning Variables

```python
## You can assign different types of values to variables, such as strings, integers, floating point numbers, and booleans.
```

```python
age = 19
height = 5.9
name = "Abdul Rehman"
is_student = True
```

```python
print("age:", age)
print("height:", height)
print("name:", name)
print("is_student:", is_student)
```

#### Naming Conventions for Variables

```python
## Variable names should be descriptive. They must always start with a letter or an underscore.
##Variable names can contain letters, numbers, and underscores, but must not start with a number. 
##Variable names are case sensitive.
```

##### valid declaration

```python
first_name = "Muhammad"
last_name = "Khan"
```

##### invalid declaration

```python
1age = 19  # Invalid: starts with a number
first-name = "Muhammad"  # Invalid: dash is not allowed
@name = "Muhammad"  # Invalid: special character not allowed
```

#### Case Sensitivity

```python
name = "A.Rehman"
Name = "Rais"
print(name == Name)  # Output: False
```

#### Understanding Variable Types

```python
age = 19  # int
height = 5.9  # float
name = "AR"  # str
is_student = True  # bool
```

```python
print(type(age))  
print(type(name))  
```

#### Type casting

```python
# You can also convert a string containing a number to an integer.(type-casting)
age = "19"
print(type(age))  # str
age_int = int(age)
print(type(age_int))  # int
```

```python
# If you try to convert a non-numeric string to an integer, you will get an error.
name = "AR"
int(name)  # This will raise a ValueError
```

#### Dynamic Typing in Python

```python
var = 10
print(var, type(var)) 
var = "Hello"
print(var, type(var))  
var = 3.14
print(var, type(var))  
```

#### User Input in Python

```python
age = input("What is your age? ")
print(age)
print(type(age))  # str

# To convert the input to integer
typed_age = int(input("What is your age? "))
print(type(typed_age))  # int
```

#### Simple Calculator Example

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
sum_ = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2
print("Sum:", sum_)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
